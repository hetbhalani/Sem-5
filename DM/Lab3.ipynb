### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
for i in numaric:
    print(i)
    print(f'mean = {df[i].mean()}')
    print(f'std = {df[i].std()}')
    print(f'min = {df[i].min()}')
    print(f'mode = {df[i].mode()[0]}')
    print(f'range = {df[i].max() - df[i].min()}')
    print(f'max = {df[i].max()}')
```
